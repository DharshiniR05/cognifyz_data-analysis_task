                                                        TASK 1 – TOP CUISINES

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "Dataset .xlsx"
df = pd.read_excel(file_path)

print("Dataset Loaded Successfully!")
print("Total Restaurants:", len(df))
```

```python
cuisines = df['Cuisines'].dropna().str.split(', ')
cuisines = cuisines.explode()

print("Cuisine Data Prepared")
```

```python
top3 = cuisines.value_counts().head(3)

print("Top 3 Cuisines:")
print(top3)
```

```python
percentage = (top3 / len(df)) * 100

print("Percentage of Top 3 Cuisines:")
print(percentage)
```

```python
plt.figure()
top3.plot(kind='bar')
plt.title("Top 3 Cuisines")
plt.xlabel("Cuisine")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.show()
```

                                                      TASK 2 – CITY ANALYSIS

```python
city_count = df['City'].value_counts()

top_city = city_count.idxmax()

print("City with Highest Number of Restaurants:")
print(top_city)
```

```python
city_avg_rating = df.groupby('City')['Aggregate rating'].mean()

print("Average Rating for Each City:")
print(city_avg_rating)
```

```python
best_city = city_avg_rating.idxmax()

print("City with Highest Average Rating:")
print(best_city)
```

                                                     TASK 3 – PRICE RANGE DISTRIBUTION

```python
price_count = df['Price range'].value_counts().sort_index()

print("Price Range Distribution:")
print(price_count)
```

```python
price_percentage = (price_count / len(df)) * 100

print("Percentage of Restaurants in Each Price Range:")
print(price_percentage)
```

```python
plt.figure()
price_count.plot(kind='bar')
plt.title("Price Range Distribution")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.show()
```

                                                TASK 4 – ONLINE DELIVERY ANALYSIS

```python
online_count = df['Has Online delivery'].value_counts()

print("Online Delivery Count:")
print(online_count)
```

```python
online_percentage = (online_count / len(df)) * 100

print("Online Delivery Percentage:")
print(online_percentage)
```

```python
rating_compare = df.groupby('Has Online delivery')['Aggregate rating'].mean()

print("Average Rating Comparison:")
print(rating_compare)
```

```python
plt.figure()
rating_compare.plot(kind='bar')
plt.title("Average Rating: Online vs No Online Delivery")
plt.xlabel("Online Delivery")
plt.ylabel("Average Rating")
plt.show()
```
